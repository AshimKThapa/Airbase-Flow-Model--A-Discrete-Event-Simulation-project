```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "doe_results.csv"
results = pd.read_csv(file)
```

```python
results['total_delays'] = results['Deterrence Mission Delays'] + results['Escort Mission Delays'] + results['Recon Mission Delays']
```

```python
results
```

```python
cases = results.groupby(['No. of bombers', 'No. of fighters', 'No. of UAVs']).agg(['std', 'mean'])['total_delays']
cases
```

```python
cases['mean'].plot( yerr=cases['std'])
```
